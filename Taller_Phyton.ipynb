11. Fraccione los datos de latitud en 4 cuadrantes y etiquete los puntos como (I,II,II y IV) según el cuadrante y cree esta nueva variable(cuadrante como tipo factor o cadena) y calcule las estadísticas descriptivas por cuadrante

```python
rd.seed (1000518135)
import numpy as np
Lonmed = np.mean(df1['Lon'])
Latmed = np.mean(df1['Lat'])
Cuadrante = []
for i in range(len(df1['Lon'])):
    if df1['Lon'][i] < Lonmed and df1['Lat'][i] < Latmed:
        Cuadrante.append('III')
    elif df1['Lon'][i] > Lonmed and df1['Lat'][i] < Latmed:
        Cuadrante.append('II')
    elif df1['Lon'][i] < Lonmed and df1['Lat'][i] > Latmed:
        Cuadrante.append('I')
    elif df1['Lon'][i] > Lonmed and df1['Lat'][i] > Latmed:
        Cuadrante.append('IV')
    else:break
df3 = pd.DataFrame({'Cuadrante':Cuadrante, 'Lon':longitude, 
                    'Lat':latitude,
                    'SMI':SMI,'NDVI':NDVI,'LST':LST})
print(df3)

df3.groupby('Cuadrante').describe()

# Para ls variables SMI, NDVI y LST en el cuadrante 3 tienden a ser mayores, sin embargo los datos no tienen mucha variacion entre cuadrantes.
```
